```python
import requests

def count_pokemon():
    url = "https://pokeapi.co/api/v2/pokemon/"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        count = data['count']
        print(f"Total number of Pokémon available: {count}")
    else:
        print("Failed to fetch data. Status code:", response.status_code)

if __name__ == "__main__":
    count_pokemon()
```

```python
import requests
import pandas as pd
import time

# Counter variable to track API calls
api_call_counter = 0

def fetch_all_pokemon_data():
    global api_call_counter  # Access the global counter variable
    
    all_pokemon_data = []
    url = "https://pokeapi.co/api/v2/pokemon/"
    
    while url:
        response = requests.get(url)
        
        # Increment the counter each time a request is made
        api_call_counter += 1
        
        if response.status_code == 200:
            data = response.json()
            for pokemon in data['results']:
                pokemon_url = pokemon['url']
                pokemon_data = requests.get(pokemon_url).json()
                all_pokemon_data.append(pokemon_data)
                
                # Print update message for each new Pokémon data obtained
                print(f"Fetched data for {pokemon_data['name'].capitalize()} from API.")
                
            url = data['next']  # Get URL for next page
        else:
            print("Failed to fetch data. Status code:", response.status_code)
            return None
    
    return all_pokemon_data

def main():
    start_time = time.time()
    
    all_pokemon_data = fetch_all_pokemon_data()
    
    end_time = time.time()
    
    if all_pokemon_data:
        # Extracting field names
        fields = []
        for pokemon in all_pokemon_data:
            for field in pokemon.keys():
                if field not in fields:
                    fields.append(field)
        
        # Create pandas DataFrame
        df = pd.DataFrame(columns=fields)
        
        # Populate DataFrame
        for pokemon in all_pokemon_data:
            df = df.append(pokemon, ignore_index=True)
        
        # Save DataFrame to CSV
        df.to_csv('pokemon_data.csv', index=False)
        print("CSV file saved successfully.")
        
        print("Total execution time:", end_time - start_time, "seconds")
        print("Total API calls made:", api_call_counter)

if __name__ == "__main__":
    main()
```
